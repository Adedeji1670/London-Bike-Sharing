```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dt_bike = pd.read_csv('London_bike.csv')
```

- t1= real temperature in C
- t2=temperature in C 'feel like'
- hum= humidity.
- is_holiday; 1=is holiday, 0=is not holiday.

#### dt_bike['is_weekend'].value_counts()
- 1 for is weekend
- 0 for is not weekend

### season =0-spring, 1=summer,2-fall, 3-winter

#### Here comes the Data cleaning and creating another columns important for this project

```python
dt_bike['timestamp'] = pd.to_datetime(dt_bike['timestamp'])
```

```python
dt_bike['year'] = dt_bike['timestamp'].dt.year
dt_bike['month'] = dt_bike['timestamp'].dt.month
```

```python
dt_bike['Time'] = dt_bike['timestamp'].dt.time
```

```python
dt_bike.head(5)
```

```python
dt_bike['monthname'] = dt_bike['timestamp'].dt.strftime('%b')
```

```python
dt_bike.head(2)
```

```python
dt_bike['days'] = dt_bike['timestamp'].dt.day_name()
```

```python
dt_bike.head(4)
```

```python
dt_bike.info()
```

```python
dt_bike.describe()
```

```python
month_order =['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
days_order =['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
```

#### HERE COMES THE DESCRIPTIVE ANALYSIS

```python
# COUNT OF BIKE share OVER TIME.
COUNT_BY_YEAR = dt_bike.groupby('year')['cnt'].sum()
print(COUNT_BY_YEAR.sort_values(ascending=False))
```

```python
# COUNT OF BIKE share OVER SEASON.
COUNT_BY_SEASON = dt_bike.groupby('season')['cnt'].sum()
print(COUNT_BY_SEASON.sort_values(ascending=False))
```

```python
# COUNT OF BIKE share OVER WEATHER.
COUNT_BY_WEATHER = dt_bike.groupby('weather_code')['cnt'].sum()
print(COUNT_BY_WEATHER.sort_values(ascending=False))
```

```python
# COUNT OF BIKE share OVER MONTHS.
COUNT_BY_MONTHS = dt_bike.groupby('monthname')['cnt'].sum()
print(COUNT_BY_MONTHS.sort_values(ascending=False))
```

```python
# COUNT OF BIKE share OVER DAYS.
COUNT_BY_DAYS = dt_bike.groupby('days')['cnt'].sum()
print(COUNT_BY_DAYS.sort_values(ascending=False))
```

```python
# COUNT OF BIKE share OVER Time.
TIME = dt_bike.groupby('Time')['cnt'].sum()
print(TIME.sort_values(ascending=False))
```

```python
# COUNT OVER TIME
plt.figure(figsize=(6,3))
COUNT_BY_YEAR.plot(kind = 'bar',color='indigo')
plt.title('COUNT BIKE BY YEAR')
plt.xlabel('year')
plt.ylabel('cnt')
plt.show()
```

#### INSIGHTS.
- The counts of bike shared increases in year 2015 and 2016 but there is a very low bike shared in 2017.
   - REASONS: We don't know maybe, the company polices changed.

```python
COUNT_BY_MONTHS= COUNT_BY_MONTHS.reindex(month_order)
```

```python
# COUNT OF BIKE SHARE OVER MONTHS
plt.figure(figsize=(8,3))
COUNT_BY_MONTHS.plot(kind = 'line',marker='o',color='indigo')
plt.title('COUNT BIKE SHARED BY MONTHS')
plt.xticks(np.arange(len(month_order)),month_order)
plt.show()
```

#### INSIGHTS.
- The count of bike shared in JANUARY,FEB,NOV and DEC is low. While there is increase in other months and a very high increase in JUL.

```python
COUNT_BY_DAYS= COUNT_BY_DAYS.reindex(days_order)
```

```python
# COUNT OF BIKE SHARE OVER DAYS
plt.figure(figsize=(8,3))
COUNT_BY_DAYS.plot(kind = 'line', marker='o',color='indigo')
plt.title('COUNT OF BIKE SHARED OVER DAYS')
plt.xlabel('year')
plt.ylabel('DAYS')
plt.xticks(np.arange(len(days_order)),days_order)
plt.show()
```

#### INSIGHTS.
- Bike shared increase during the working days while it decreases during the weekend(SATURDAY and SUNDAY)
- Maybe people used it most for going and coming to work

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12,4))
sns.barplot(data=dt_bike, x = 'season', y = 'cnt', ax = axes[0], color='green')
sns.barplot(data=dt_bike, x = 'is_holiday', y = 'cnt', ax = axes[1])
axes[0].set_title('COUNT OF BIKE SHARED IN A SEASON')
axes[1].set_title('BIKE SHARED DURING HOLIDAYS')
plt.suptitle('BIKE SHARED BETWEEN SEASON AND HOLIDAYS')
```

```python
dt_bike['season'].value_counts()
```

```python
dt_bike['is_holiday'] = ['NO' if x == 0.0 else 'YES' if
                     x == 1.0 else
                     x for x in dt_bike['is_holiday']]
```

```python
dt_bike['is_holiday'].value_counts()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12,4))
sns.barplot(data=dt_bike, x = 'season', y = 'cnt', ax = axes[0], color='green')
sns.barplot(data=dt_bike, x = 'is_holiday', y = 'cnt', ax = axes[1])
axes[0].set_title('COUNT OF BIKE SHARED IN A SEASON')
axes[1].set_title('BIKE SHARED DURING HOLIDAYS')
plt.suptitle('BIKE SHARED BETWEEN SEASON AND HOLIDAYS')
```

#### INSIGHTS.
- Bike shared during the holidays is very low
- Bike shared during the working days is very high, which simply means people working use it the most.

```python
dt_bike['weather_code'].value_counts()
```

```python
dt_bike['weather_code'] = ['clear' if x==1 else 'scattered clouds' if x==2 else 'broken_clouds' if x==3
                           else 'cloudy' if x==4 else 'Rain' if x==7
                           else 'Rain_with_thunderstorm' if x==10
                           else 'Snowfall' if x==26 else 'Freezing_fog' if x==94 else x for x in dt_bike['weather_code']]
```

```python
dt_bike['weather_code'].value_counts()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(15,6))
plt.xticks(rotation=45)  
sns.barplot(data=dt_bike, x = 'weather_code', y = 'cnt', ax = axes[0], color='indigo')
sns.barplot(data=dt_bike, x = 'is_weekend', y = 'cnt', ax = axes[1])
axes[0].set_title('weather vs count')
axes[1].set_title('weekend vs count')
plt.suptitle('count against weather and weekend')
```

#### INSIGHTS.
- Bike shared during RAIN WITH THUNDERSTOME, SNOWFALL is very low.
- Bike shared during SCATTERED CLOUDS,BROKEN CLOUDS and CLEAR is high.

```python
plt.figure(figsize=(14,8))
ax = sns.barplot(x=dt_bike['season'],data=dt_bike, palette='muted', y='cnt', hue=dt_bike['weather_code'])
plt.title('Bike share on weekdays and weekends')
plt.xlabel('season')
plt.ylabel('Bike shares')
```

#### INSIGHTS.
- Regardless of the weather code bike was shared every season, just that the rate differs.
